# ***HISTOGRAM***

```python
sns.set(rc={'figure.figsize' : (12,8)})
sns.displot(data['CGPA'], kde=False, bins=20);
```

> *Here we can observe that the total number of records based on gender=2966in those two are unique ,they are male and female.
Male records are higher than Female and it's frequency is 2475.*

```python
data.hist(edgecolor='black',bins = 25, figsize= (15,10))
plt.show()
```

# *JOINTPLOT*

```python
sns.jointplot(x="Gender", y="Stream", data=data, height=5)
plt.show()
```

> *From the above plot we conclude that the male and female are more in Computer Science and Information Technology.*

# *SCATTERPLOT*

```python
sns.scatterplot(x='Age', y='Internships', data=data)
```

> *From the above plot it shows that the more internships have got in between the age of 20 to 26 and in between 28 to 30 they got less internships*.

# *BOXPLOT*

```python
sns.boxplot(x="Stream", y="HistoryOfBacklogs", data=data)
plt.show()
```

> *The above plot represents that the history of backlogs of people are having in only one subject i.e. in electronics and communication.*

# *violinplot*

```python
sns.violinplot(x="Stream", y="Age",hue="PlacedOrNot",  data=data,split=True)
```

> *by the above violinplot we can predict the placements based on stream and age.*

```python
sns.violinplot(x ="Gender", y ="Internships", hue ="PlacedOrNot",
data = data, split = True)
```

> *by the above violinplot we can predict the placements based on gender and no of internships*
